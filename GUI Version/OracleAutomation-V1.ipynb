```python
# pypy/TKinterGui/GUI Version/OracleAutomation-V1.ipynb

import tkinter as tk
from tkinter import messagebox
import pyautogui as pg
import time
import subprocess

def open_firefox_dedicated_browser(shortcut_path):
    try:
        subprocess.Popen([shortcut_path], shell=True)
        print("Mozilla Firefox is now opening...")
    except Exception as e:
        print(f"Error: {e}")


def user_log_in(username, password):
    # Load the webpage
    time.sleep(8)

    # Login the website
    pg.write(username)
    pg.press("tab")
    pg.write(password)
    pg.press("tab")
    pg.press("enter")

    # Wait for the page to load 
    time.sleep(2)
# -------------------------------------------------------------------------------------
def log_in():
    root.withdraw()  # Minimize the Tkinter window
    open_firefox_dedicated_browser(firefox_shortcut_path)
    user_log_in("5638", "ripon")
    root.deiconify()  # Show the Tkinter window after the login process completes


def download_OHD():
    root.withdraw()  # Minimize the Tkinter window
    
    # Wait for the page to load 
    time.sleep(2)
    
    # Select the web element for opening oracle software for creating batch
    pg.press("tab", presses=14)
    pg.press("enter")
    time.sleep(1)
    pg.press("tab", presses=3)
    pg.press("enter")
    
    # Wait for the software download 
    time.sleep(8)
    pg.hotkey("alt", "i")
    pg.press("esc")
    pg.press("enter")
    
    # Check On Hand Details
    time.sleep(8)
    pg.hotkey("alt", "v")
    pg.press("r")
    pg.hotkey("alt", "n")
    time.sleep(5)
    pg.write("S")
    pg.press("tab")
    time.sleep(5)
    pg.press("down", presses=5)
    time.sleep(2)
    pg.press("enter")
    time.sleep(1)
    pg.hotkey("ctrl", "c")
    pg.press("tab", presses=3)
    pg.press("enter")
    time.sleep(3)
    pg.hotkey("alt", "m")
    time.sleep(3)
    pg.hotkey("alt", "n")
    time.sleep(3)
    pg.hotkey("alt", "i")
    time.sleep(3)
    pg.hotkey("alt", "r")
    time.sleep(2)
    pg.hotkey("alt", "p")
    
    print("Downloading the On Hand Details ")

    root.deiconify()  # Show the Tkinter window after the login process completes
    



def create_mo():
    messagebox.showinfo("Create MO", "Creating MO...")

def create_batch_posting():
    messagebox.showinfo("Create Batch Posting", "Creating Batch Posting...")

def check_batch_report():
    messagebox.showinfo("Check Batch Report", "Checking Batch Report...")

def check_ohd():
    download_OHD()
    messagebox.showinfo("Check OHD", "Checking OHD...")

def check_ils():
    messagebox.showinfo("Check ILS", "Checking ILS...")

def check_m_transaction():
    messagebox.showinfo("Check M_Transaction", "Checking M_Transaction...")



def create_gui():
    def button_click(button_name):
        if button_name == "Create MO":
            create_mo()
        elif button_name == "Create Batch Posting":
            create_batch_posting()
        elif button_name == "Check Batch Report":
            check_batch_report()
        elif button_name == "Check OHD":
            check_ohd()
        elif button_name == "Check ILS":
            check_ils()
        elif button_name == "Check M_Transaction":
            check_m_transaction()
        elif button_name == "Log In":
            log_in()

    buttons = [
        "Log In",
        "Check OHD",
    ]

    for button_name in buttons:
        button = tk.Button(root, text=button_name, command=lambda btn=button_name: button_click(btn))
        button.pack(pady=5)

if __name__ == "__main__":
    firefox_shortcut_path = r"C:\ProgramData\Microsoft\Windows\Start Menu\Programs\Mozilla Firefox.lnk"
    root = tk.Tk()
    root.title("Code Run")
    create_gui()
    root.mainloop()
```
